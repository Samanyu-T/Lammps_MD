```python
pe = []
var = []
for file in glob.glob('Monte_Carlo_HSurface_2/*/N_species.npy'):
    file_var = file.split('/')[1].split('_')

    var.append([float(file_var[1]), float(file_var[3]), int(file_var[5])])
    pe.append(np.loadtxt(file)[:,1])
    
pe = np.array(pe)
var = np.array(var)
```

```python
idx = np.lexsort( (var[:,2], var[:,1], var[:,0]) )

var = var[idx]

pe = pe[idx]

temp_unique = np.unique(var[:,1])

print(temp_unique)
```

```python
data = {}

for temp in temp_unique:

    t_idx = np.where(var[:,1] == temp)[0]
    
    key = 'Temp:%d' % temp

    data[key] = {}

    data[key]['x'] = np.arange(len(pe[0,:]))

    for i, idx in enumerate(t_idx):
        
        rep_key = 'Replica:%d' % i

        data[key][rep_key] = pe[idx,:]

df = pd.DataFrame(data)
```
